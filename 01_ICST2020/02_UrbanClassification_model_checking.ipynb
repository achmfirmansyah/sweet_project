```
lokus=['jabodetabek','mebidangro','gerbangkertasusila','kedungsepur','maminasata','bandungraya']
```

```
from sklearn import preprocessing
```

```
C_range = [0.1,1,10,100]
param_grid = dict( C=C_range)
string_grid=[]
for lokasi in lokus:
  print(lokasi)
  dataset = rasterio.open('/content/gdrive/My Drive/Urban_monitoring/Urban_/compiled_GHSL_30_train_2014_'+lokasi+'.tif')
  dataset_bands=pd.DataFrame()
  for i in dataset.indexes:
    temp=dataset.read(i)
    temp=pd.DataFrame(data=np.array(temp).flatten()).rename(columns={0:i})
    dataset_bands=temp.join(dataset_bands)
  dataset_bands.rename(columns={1:'BU_class',2:'NDVI_landsat7',3:'NDBI_landsat7',4:'MNDWI_landsat7',
                            5:'SAVI_landsat7',6:'LSE_landsat7',7:'rad_VIIRS'},inplace=True)
  dataset_bands['U_class']=dataset_bands.BU_class.apply(lambda y: 1 if y>=3 else 0)
  dataset_bands['LSE_der_landsat7']=dataset_bands.NDVI_landsat7.apply(lambda y: 0.995 if y < -0.185 else (
    0.970 if y< 0.157 else (1.0098+0.047*np.log(y) if y<0.727 else 0.990)))
  dataset=dataset_bands.query('U_class==0').sample(10000).append(dataset_bands.query('U_class==1').sample(10000))
  dataset=dataset.reset_index()[['rad_VIIRS','SAVI_landsat7','MNDWI_landsat7','NDBI_landsat7','NDVI_landsat7','LSE_der_landsat7','U_class']]
  svclassifier = SVC(kernel='rbf')
  minmax_=preprocessing.MinMaxScaler()
  X=minmax_.fit_transform(dataset[['rad_VIIRS','SAVI_landsat7','NDVI_landsat7','MNDWI_landsat7','NDBI_landsat7','LSE_der_landsat7']])
  y=dataset[['U_class']]
  cv = StratifiedShuffleSplit(n_splits=10, test_size=0.2, random_state=42)
  grid = GridSearchCV(SVC(), param_grid=param_grid, cv=cv,verbose=1)
  grid.fit(X, y)
  str_r="The best parameters are %s %s with a score of %0.2f"% (grid.best_params_,lokasi, grid.best_score_)
  string_grid.append(str_r)
```

```
from sklearn.metrics import classification_report, confusion_matrix
dataset = rasterio.open('/content/gdrive/My Drive/Urban_monitoring/Urban_/compiled_GHSL_30_train_2014_jabodetabek.tif')
dataset_bands=pd.DataFrame()
for i in dataset.indexes:
  temp=dataset.read(i)
  temp=pd.DataFrame(data=np.array(temp).flatten()).rename(columns={0:i})
  dataset_bands=temp.join(dataset_bands)
dataset_bands.rename(columns={1:'BU_class',2:'NDVI_landsat7',3:'NDBI_landsat7',4:'MNDWI_landsat7',
                            5:'SAVI_landsat7',6:'LSE_landsat7',7:'rad_VIIRS'},inplace=True)
dataset_bands['U_class']=dataset_bands.BU_class.apply(lambda y: 1 if y>=3 else 0)
dataset_bands['LSE_der_landsat7']=dataset_bands.NDVI_landsat7.apply(lambda y: 0.995 if y < -0.185 else (
    0.970 if y< 0.157 else (1.0098+0.047*np.log(y) if y<0.727 else 0.990)))
dataset=dataset_bands.query('U_class==0').sample(10000).append(dataset_bands.query('U_class==1').sample(10000))
minmax_=preprocessing.MinMaxScaler()
dataset=dataset.reset_index()[['rad_VIIRS','SAVI_landsat7','MNDWI_landsat7','NDBI_landsat7','NDVI_landsat7','LSE_der_landsat7','U_class']]  
X_train, X_test, y_train, y_test = train_test_split(dataset[['rad_VIIRS','SAVI_landsat7','MNDWI_landsat7','NDVI_landsat7','NDBI_landsat7','LSE_der_landsat7']],
                                                    dataset[['U_class']], test_size = 0.20)
svclassifier = SVC(kernel='rbf',C=100,random_state=123)
X_train=minmax_.fit_transform(X_train)
X_test=minmax_.fit_transform(X_test)
svclassifier.fit(X_train, y_train)
y_pred = svclassifier.predict(X_test)
from sklearn.metrics import classification_report, confusion_matrix
print(confusion_matrix(y_test, y_pred))
print(classification_report(y_test, y_pred))
rfclassifier = RandomForestClassifier(random_state=123,max_depth=6,n_estimators=500,criterion='entropy')
rfclassifier.fit(X_train,y_train)
y_pred=rfclassifier.predict(X_test)
print(confusion_matrix(y_test, y_pred))
print(classification_report(y_test, y_pred))
xgclassifier = xgb.XGBClassifier(random_state=123,colsample_bytree= 0.7, learning_rate=0.07, max_depth= 4, 
                             min_child_weight=11, n_estimators= 500, nthread= 4, objective= 'binary:logistic', 
                             seed= 123, silent=1, subsample= 0.8)
xgclassifier.fit(X_train,y_train)
y_pred=xgclassifier.predict(X_test)
print(confusion_matrix(y_test, y_pred))
print(classification_report(y_test, y_pred))
from sklearn.model_selection import KFold,cross_val_predict,cross_val_score,cross_validate
X=dataset[['rad_VIIRS','SAVI_landsat7','MNDWI_landsat7','NDVI_landsat7','NDBI_landsat7','LSE_der_landsat7']]
y=dataset[['U_class']]
svc_acc_jabodetabek=np.mean(cross_val_score(svclassifier, X, y, cv=10,scoring='roc_auc'))
rbf_acc_jabodetabek=np.mean(cross_val_score(rfclassifier, X, y, cv=10,scoring='roc_auc'))
xgb_acc_jabodetabek=np.mean(cross_val_score(xgclassifier, X, y, cv=10,scoring='roc_auc'))
```
